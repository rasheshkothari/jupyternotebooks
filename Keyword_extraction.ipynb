# BEER NAME EXTRACTION

```python
def extract_beer(text, list_of_keywords):    
    res = ""
    text = text.split(" ")
    for ber in list_of_keywords:
        if ber in text:
            idx = text.index(ber)
            res = " ".join(text[:idx])
            break
        else:
            continue
    return res
```

```python
# Example
extract_beer('MILLER HIGH LIFE BEER 6 CT 7 OZ GLASS BOTTLE IN CRTN 42 OZ',['BEER'])
```

```python
list_of_keywords = ["BEER"]
for i in PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION.index:
#     print(type(i))
    beer = PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION.loc[i, "PRODUCT_DESCRIPTION"]
    if type(beer) == str :
        PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION.loc[i, 'beer_name'] = extract_beer(beer,list_of_keywords)
    else: 
        PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION.loc[i, 'beer_name'] = " "
```

# Extract pattern: BOTTLE 

```python
def extract_bottle(text, list_of_keywords):    
    res = ""
    text = text.split(" ")
    for btl in list_of_keywords:
        if btl in text:
            idx = text.index(btl)
            res = " ".join(text[idx:])
            break
        else:
            continue
    return res
```

```python
list_of_keywords =['BOTTLE', 'BTL','BOTLE', 'BTTL','BTLE', 'BOTTLES']
for i in PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION[:300].index:
    bottle = PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION.loc[i, "PRODUCT_DESCRIPTION"]
    if type(bottle) == str :
        PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION.loc[i, 'bottle'] = extract_bottle(bottle,list_of_keywords)
    else: 
        PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION.loc[i, 'bottle'] = " "
```

# PATTERN N CT N OZ

```python
def pattern_n_ct_n_oz(description):
    return description.str.extract(r'(\d\s+CT\s+\d\s+OZ)', expand=False).str.strip()
```

```python
PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION['ctoz'] = pattern_n_ct_n_oz(PRODUCT_MASTER_CATALOG_MAPPED_TEMP_PRODUCT_DESCRIPTION['PRODUCT_DESCRIPTION'])
```
